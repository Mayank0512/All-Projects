# DATA PREPARTION

Choosing the Top 10 classes from classification

```
# count numbers of instances per class
from collections import Counter
cnt = Counter(df.classification)
# select only K most common classes! - was 10 by default
top_classes = 10
# sort classes
sorted_classes = cnt.most_common()[:top_classes]
classes = [c[0] for c in sorted_classes]
counts = [c[1] for c in sorted_classes]
print("at least " + str(counts[-1]) + " instances per class")

# apply to dataframe
print(str(df.shape[0]) + " instances before")
df = df[[c in classes for c in df.classification]]
print(str(df.shape[0]) + " instances after")

seqs = df['merge_col']
lengths = [len(s) for s in seqs]
```

```
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y = lb.fit_transform(df.classification)
```

```
X=df.drop('classification',axis=1)
```

```
print(X.shape)
print(y.shape)
```

```
from keras.preprocessing import text, sequence
from keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split

max_length = max(lengths)

tokenizer = Tokenizer(char_level=True)
tokenizer.fit_on_texts(seqs)

X_seq = tokenizer.texts_to_sequences(seqs)
X_seq = sequence.pad_sequences(X_seq, maxlen=max_length)
```

# CREATING THE MODEL

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from keras.models import Model
from keras.layers.embeddings import Embedding
from tensorflow.keras.layers import Dense,Conv1D,MaxPooling1D,LSTM,Flatten,Input
```

```
input_1 = Input(shape=(max_length,)) 
```

```
embedding_layer = Embedding(len(tokenizer.word_index)+1, 8)(input_1)
CNN_Layer_1 = Conv1D(filters=64,kernel_size=6,padding='same',activation='relu')(embedding_layer)
MaxPooling_layer_1 = MaxPooling1D(pool_size=2)(CNN_Layer_1)
CNN_Layer_2 = Conv1D(filters=32,kernel_size=3,padding='same',activation='relu')(MaxPooling_layer_1)
MaxPooling_layer_2 = MaxPooling1D(pool_size=2)(CNN_Layer_2)
LSTM_Layer_1=Flatten()(MaxPooling_layer_2)
#LSTM_Layer_1 = LSTM(150)(MaxPooling_layer_2)
Dense_1=Dense(128,activation='relu')(LSTM_Layer_1)
output = Dense(top_classes, activation='softmax')(Dense_1)
model_2 = Model(inputs=input_1, outputs=output)
```

```
from tensorflow.keras.optimizers import Adam
opt = Adam(learning_rate=0.001)
model_2.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['acc'])
print(model_2.summary())
```

## TRAINING AND TESTING THE MODEL

```
X_seq = np.array(X_seq)
X_seq = np.array(list(X_seq))
X_seq_train,X_seq_test,y_train,y_test=train_test_split(X_seq,y,test_size=0.3,random_state=42)
```

```
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
```

```
early_stop = EarlyStopping(monitor='val_loss',patience=5)
```

```
best_model_save_2 = ModelCheckpoint(
    'Protein Classification_2.h5', monitor='val_acc', verbose=1, save_best_only=True,
    save_weights_only=False, mode='auto'
)
```

```
history = model_2.fit(x=X_seq_train ,y=y_train, batch_size=128, epochs=15, verbose=1, validation_split=0.2,callbacks=[best_model_save_2])
```

```
best_model_2 = tf.keras.models.load_model('Protein Classification_2.h5')
```

```
from sklearn.metrics import confusion_matrix
y_test_pred = best_model_2.predict(X_seq_test)
confusionMatrix_2 = confusion_matrix(np.argmax(y_test,axis=1), np.argmax(y_test_pred,axis=1))
print(confusionMatrix_2)
```
